```python
import pandas as pd

df = pd.read_csv("train_FD001.txt", sep="\s+", header=None)

columns = ['unit', 'cycle',
           'op1', 'op2', 'op3'] + \
          [f'sensor{i}' for i in range(1, 22)]

df.columns = columns

```

```python
df.head(20)
```

```python
print("Number of engines:", df['unit'].nunique())
```

```python
max_cycles = df.groupby('unit')['cycle'].max()

print("First 10 engines max cycles:")
print(max_cycles.head(10))
```

```python
sensor_cols = [col for col in df.columns if 'sensor' in col]
print("Number of sensors:", len(sensor_cols))
```

```python
print("Missing values per column:")
print(df.isnull().sum())
```
